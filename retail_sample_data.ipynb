# Import libraries

```python
import pandas as pd
import numpy as np
```

# Load the dataset

```python
# prompt: read csv file from local computer

from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

data = pd.read_csv(fn)
data
```

# 1. Handling Missing Values
# Drop rows where "Name" and "Email" are missing

```python
data = data.dropna(subset=['Name', 'Email'])
data
```

# Impute missing "Age" with the median age

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data
```

# Drop rows where "Purchase_Date" is missing

```python
data = data.dropna(subset=['Purchase_Date'])
data
```

# 2. Correcting Data Format
# Ensure consistent date format for "Purchase_Date"

```python
data['Purchase_Date'] = pd.to_datetime(data['Purchase_Date'], errors='coerce')
data
```

# Standardize categorical data

```python
data['Payment_Method'] = data['Payment_Method'].str.title()
data['Country'] = data['Country'].str.title()
data
```

# 3. Handling Outliers
# Remove rows with negative prices or unusual values

```python
data = data[data['Price'] >= 0]
data['Total_Amount'] = data['Quantity'] * data['Price']
data
```

# 4. Data Validation
# Verify "Total_Amount" is calculated correctly

```python
data['Total_Amount_Valid'] = data['Quantity'] * data['Price']
data = data[data['Total_Amount'] == data['Total_Amount_Valid']]
data
```

# Drop the helper column used for validation

```python
data = data.drop(columns=['Total_Amount_Valid'])
data
```

# 5. Duplicate Removal
# Remove duplicate rows

```python
data = data.drop_duplicates()
data
```

# 6. Normalization (Optional)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['Age', 'Price', 'Total_Amount']] = scaler.fit_transform(data[['Age', 'Price', 'Total_Amount']])
data
```

# 7. Export Cleaned Data

```python
data.to_csv('cleaned_customer_purchases.csv', index=False)

print("Data cleaning complete. Cleaned data saved to 'cleaned_customer_purchases.csv'")
```
