```python
import pandas as pd
import numpy as np

# prompt: read csv file from local computer

from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

data = pd.read_csv(fn)
data
```

```python
# prompt: Drop rows where "Name" and "Email" are missing
# Impute missing "Age" with the median age
# Drop rows where "Purchase_Date" is missing
# Ensure consistent date format for "Purchase_Date"
# Standardize categorical data
# Remove rows with negative prices or unusual values
# Verify "Total_Amount" is calculated correctly

# Drop rows where "Name" and "Email" are missing
data.dropna(subset=['Name', 'Email'], inplace=True)

# Impute missing "Age" with the median age
data['Age'].fillna(data['Age'].median(), inplace=True)

# Drop rows where "Purchase_Date" is missing
data.dropna(subset=['Purchase_Date'], inplace=True)

# Ensure consistent date format for "Purchase_Date"
data['Purchase_Date'] = pd.to_datetime(data['Purchase_Date'], errors='coerce')
data.dropna(subset=['Purchase_Date'], inplace=True)

# Standardize categorical data (example: converting to lowercase)
for column in data.select_dtypes(include=['object']):
    data[column] = data[column].str.lower()

# Remove rows with negative prices or unusual values (adjust based on your data)
data = data[data['Price'] >= 0]

# Verify "Total_Amount" is calculated correctly (if applicable)
# Assuming Total_Amount = Price * Quantity, you can check and correct it:
if 'Quantity' in data.columns and 'Total_Amount' in data.columns:
    data['Total_Amount'] = data['Price'] * data['Quantity']

# Display the cleaned dataset
data

```
